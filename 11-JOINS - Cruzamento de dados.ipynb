```python
#Cria sessão Spark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import pyspark.sql.functions as F
```

```python
#Instância Spark
spark = SparkSession.builder.appName('Curso Pyspark') \
         .config('spark.sql.repl.eagerEval.enabled', True) \
         .getOrCreate()      
```

```python
# lendo um arquivo parquet
autores = spark.read.parquet('./DATASETS/AUTORES.parquet')
livros = spark.read.parquet('./DATASETS/LIVROS.parquet')
clientes = spark.read.parquet('./DATASETS/CLIENTES.parquet')
compras = spark.read.parquet('./DATASETS/COMPRAS.parquet')
```

```python
autores.show(20, False)
```

```python
livros.show(20, False)
```

```python
# fazendo um left join da tabela autores e livros
autores.join(other=livros, on='id', how='left').show(20, False)
```

```python
# fazendo um right join da tabela autores e livros
autores.join(other=livros, on='id', how='right').show(20, False)
```

```python
# iner jopin sào registros que pertencem as duas bases
autores.join(other=livros, on='id', how='inner').show(20, False)
```

```python
# iner jopin sào registros que pertencem as duas bases
# autores.join(other=livros, on=['id', 'segunda_coluna'], how='inner').show(20, False)
```

```python
compras.show(20, False)
```

```python
compras.join(other=livros, on=compras.cd_livro == livros.id, how='inner').show(20, False)
```

```python
# pode ser feito sem mencionar o nome das variaveis, somente seguindo a ordem
compras.join(livros, compras.cd_livro == livros.id, 'inner').show(20, False)
```

```python
# cond = [compras.cd_livro == livros.id, livros.preco < 20, compras.cartao_bandeira  == 'American Express']
cond = [(compras.cd_livro == livros.id) & (livros.preco < 20) | (compras.cartao_bandeira  == 'American Express')]
```

```python
# também pode incluir novos filtros
compras.join(livros, cond, 'inner').show(20, False)
```

```python
df1 = autores.where(autores.id.isin('106042959', '017551704', '054099410'))
df1.show(truncate=0)

```

```python
df2 = livros.where(livros.id.isin('017551704'))
df2.show(truncate=0)
```

```python
# inner join
# pertence aos dois ao mesmo tempo
df1.join(df2, 'id', 'inner').show(truncate=0)
```

```python
# left join
# tudo do df1 e o que foi encontrado no df 2
df1.join(df2, 'id', 'left').show(truncate=0)
```

```python
# right join
# tudo do df2 e o que foi encontrado no df 1
df1.join(df2, 'id', 'right').show(truncate=0)
```

```python
df3 = livros.where(livros.id.isin('028125446', '017551704', '049806867'))
df3
```

```python
# full join
df1.join(df3, 'id', 'full')
```

```python
# cross join
# duplicar todos os registros cruzando os dados do df1 com o df3
df1.join(df3, how='cross').orderBy('titulo')
```

```python
# anti join
# só mostras os registros excluisivos do df1
df1.join(df3, 'id', 'anti')
```

```python
df4 = df3.union(df3)
df4
```

```python
# semi join
# apenas registros que tem no df1 e e df4 removendo duplicidade
df1.join(df4, 'id', 'semi')
```
