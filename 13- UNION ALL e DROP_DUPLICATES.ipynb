```python
#Cria sessão Spark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import pyspark.sql.functions as F
```

```python
#Instância Spark
spark = SparkSession.builder.appName('Curso Pyspark') \
         .config('spark.sql.repl.eagerEval.enabled', True) \
         .getOrCreate()      
```

```python
# lendo um arquivo parquet
df = spark.read.parquet('./DATASETS/LIVROS.parquet').limit(5)
```

```python
df.show(truncate=0)
```

```python
# union precisa que os dfs estejam na mesma estrutura
dup = df.union(df)
dup.show(truncate=0)
```

```python
df1 = df.select('cnpj_editora', 'data_lancamento', 'ean', 'isbn10', 'numero_paginas', 'preco', 'id')
df1.show()
```

```python
df.union(df1)
```

```python
# só impilha as colunas que existem nos dois dataframes
df2 = df1.withColumn('quantidade', F.lit(1))
df.unionByName(df1)
```

```python
# allowMissingColumns = True permite colunas nulas

df2.unionByName(df1, allowMissingColumns=True)
```

```python
# dados duplicados
dup
```

```python
# remover dados duplicados

dup.dropDuplicates()

# pode ser assim: dup.drop_duplicates()

```
